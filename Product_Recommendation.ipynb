# 3.1 Logistic Regression

```python
# Resampling the data to get a suffled dataset
df=df.sample(frac=1).copy()

# Considering all the features except P1_Ratio as dependent features
X=df.iloc[:,1:-1].drop(["P1_Ratio"],axis=1)

# Taking P1_Ratio as target variable
y=df["P1_Ratio"]

# Splitting into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)

# Creating an instance of logistic regression model
logisticRegr = LogisticRegression(solver='lbfgs')

# Fitting the model for training samples
logisticRegr.fit(X_train, y_train)

# Predicting for test samples 
y_pred=logisticRegr.predict(X_test)

# Printing the accuracy score of the model
print("accuracy_score",accuracy_score(y_test,y_pred))

logisticRegr.score(X_test, y_test)

# Obtaining confusion matrix of the classifier
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Collecing the indices of customer companies which have participated in test set
test1=y_test.index.to_list()

# Dropping the companies which have participated in test set so that they can be replaced with fresh set of samples in test output
# A fresh set of 1386 samples are selected at random and are taken as next test set
y_test2=y.drop(test1,axis=0).sample(n=1386)

# The samples excluding from the new test set are taken into training output
y_train2=y.drop(y_test2.index.to_list(),axis=0)

# The independent features corresponding to the new training output are taken into new training set of independent features
X_train2=X.loc[y_train2.index.to_list()]

# The independent features corresponding to the new test output are taken into new test set of independent features
X_test2=X.loc[y_test2.index.to_list()]

# Fitting model with new set of training and test data
logisticRegr.fit(X_train2, y_train2)

# Predicting for target variable in test set
y_pred2=logisticRegr.predict(X_test2)

# Printing the accuracy of the model
print("accuracy_score",accuracy_score(y_test2,y_pred2))

# Printing the confusion matrix of the classifier
cm2 = confusion_matrix(y_test2, y_pred2)
print(cm2)

# The above steps are repeated till each instance in the data has appeared once in the test set

test2=y_test2.index.to_list()

test2.extend(test1)
y_test3=y.drop(test2,axis=0).sample(n=1386)

y_train3=y.drop(y_test3.index.to_list(),axis=0)

X_train3=X.loc[y_train3.index.to_list()]

X_test3=X.loc[y_test3.index.to_list()]

logisticRegr.fit(X_train3, y_train3)

y_pred3=logisticRegr.predict(X_test3)

print("accuracy_score",accuracy_score(y_test3,y_pred3))

cm3 = confusion_matrix(y_test3, y_pred3)
print(cm3)

test3=y_test3.index.to_list()

test3.extend(test2)


y_test4=y.drop(test3,axis=0).sample(n=462)

y_train4=y.drop(y_test4.index.to_list(),axis=0)

X_train4=X.loc[y_train4.index.to_list()]

X_test4=X.loc[y_test4.index.to_list()]

logisticRegr.fit(X_train4, y_train4)

y_pred4=logisticRegr.predict(X_test4)

print("accuracy_score",accuracy_score(y_test4,y_pred4))

cm4 = confusion_matrix(y_test4, y_pred4)
print(cm4)
```

<font color='green'><font size=4>We have checked all the samples in the data if they are converting from 0 to 1 after prediction using logistic regression model. Now we will collect all the indices for which P1_Ratio is converting from 0 to 1

```python
collect_index=[]
for i in range(len(y_test)):
    if (y_test.iloc[i]==0) & (y_pred[i]==1): # y_pred corresponds to first test set scenario
        collect_index.append(y_test.index[i])
for i in range(len(y_test2)):
    if (y_test2.iloc[i]==0) & (y_pred2[i]==1): # y_pred2 corresponds to second test set scenario
        collect_index.append(y_test2.index[i])
for i in range(len(y_test3)):
    if (y_test3.iloc[i]==0) & (y_pred3[i]==1): # y_pred3 corresponds to third test set scenario
        collect_index.append(y_test3.index[i])
for i in range(len(y_test4)):
    if (y_test4.iloc[i]==0) & (y_pred4[i]==1): # y_pred4 corresponds to fourth test set scenario
        collect_index.append(y_test4.index[i])
```

```python
# Resampling the data to get a suffled dataset
df=df.sample(frac=1).copy()

# Considering all the features except P2_Ratio as dependent features
X=df.iloc[:,1:-1].drop(["P2_Ratio"],axis=1)

# Taking P2_Ratio as target variable
y=df["P2_Ratio"]

# Splitting into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)

# Creating an instance of logistic regression model
logisticRegr = LogisticRegression(solver='lbfgs')

# Fitting the model for training samples
logisticRegr.fit(X_train, y_train)

# Predicting for test samples 
y_pred=logisticRegr.predict(X_test)

# Printing the accuracy score of the model
print("accuracy_score",accuracy_score(y_test,y_pred))

logisticRegr.score(X_test, y_test)

# Obtaining confusion matrix of the classifier
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Collecing the indices of customer companies which have participated in test set
test1=y_test.index.to_list()

# Dropping the companies which have participated in test set so that they can be replaced with fresh set of samples in test output
# A fresh set of 1386 samples are selected at random and are taken as next test set
y_test2=y.drop(test1,axis=0).sample(n=1386)

# The samples excluding from the new test set are taken into training output
y_train2=y.drop(y_test2.index.to_list(),axis=0)

# The independent features corresponding to the new training output are taken into new training set of independent features
X_train2=X.loc[y_train2.index.to_list()]

# The independent features corresponding to the new test output are taken into new test set of independent features
X_test2=X.loc[y_test2.index.to_list()]

# Fitting model with new set of training and test data
logisticRegr.fit(X_train2, y_train2)

# Predicting for target variable in test set
y_pred2=logisticRegr.predict(X_test2)

# Printing the accuracy of the model
print("accuracy_score",accuracy_score(y_test2,y_pred2))

# Printing the confusion matrix of the classifier
cm2 = confusion_matrix(y_test2, y_pred2)
print(cm2)

# The above steps are repeated till each instance in the data has appeared once in the test set

test2=y_test2.index.to_list()

test2.extend(test1)
y_test3=y.drop(test2,axis=0).sample(n=1386)

y_train3=y.drop(y_test3.index.to_list(),axis=0)

X_train3=X.loc[y_train3.index.to_list()]

X_test3=X.loc[y_test3.index.to_list()]

logisticRegr.fit(X_train3, y_train3)

y_pred3=logisticRegr.predict(X_test3)

print("accuracy_score",accuracy_score(y_test3,y_pred3))

cm3 = confusion_matrix(y_test3, y_pred3)
print(cm3)

test3=y_test3.index.to_list()

test3.extend(test2)


y_test4=y.drop(test3,axis=0).sample(n=462)

y_train4=y.drop(y_test4.index.to_list(),axis=0)

X_train4=X.loc[y_train4.index.to_list()]

X_test4=X.loc[y_test4.index.to_list()]

logisticRegr.fit(X_train4, y_train4)

y_pred4=logisticRegr.predict(X_test4)

print("accuracy_score",accuracy_score(y_test4,y_pred4))

cm4 = confusion_matrix(y_test4, y_pred4)
print(cm4)
```

```python
# Resampling the data to get a suffled dataset
df=df.sample(frac=1).copy()

# Considering all the features except P5_Ratio as dependent features
X=df.iloc[:,1:-1].drop(["P5_Ratio"],axis=1)

# Taking P5_Ratio as target variable
y=df["P5_Ratio"]

# Splitting into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)

# Creating an instance of logistic regression model
logisticRegr = LogisticRegression(solver='lbfgs')

# Fitting the model for training samples
logisticRegr.fit(X_train, y_train)

# Predicting for test samples 
y_pred=logisticRegr.predict(X_test)

# Printing the accuracy score of the model
print("accuracy_score",accuracy_score(y_test,y_pred))

logisticRegr.score(X_test, y_test)

# Obtaining confusion matrix of the classifier
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Collecing the indices of customer companies which have participated in test set
test1=y_test.index.to_list()

# Dropping the companies which have participated in test set so that they can be replaced with fresh set of samples in test output
# A fresh set of 1386 samples are selected at random and are taken as next test set
y_test2=y.drop(test1,axis=0).sample(n=1386)

# The samples excluding from the new test set are taken into training output
y_train2=y.drop(y_test2.index.to_list(),axis=0)

# The independent features corresponding to the new training output are taken into new training set of independent features
X_train2=X.loc[y_train2.index.to_list()]

# The independent features corresponding to the new test output are taken into new test set of independent features
X_test2=X.loc[y_test2.index.to_list()]

# Fitting model with new set of training and test data
logisticRegr.fit(X_train2, y_train2)

# Predicting for target variable in test set
y_pred2=logisticRegr.predict(X_test2)

# Printing the accuracy of the model
print("accuracy_score",accuracy_score(y_test2,y_pred2))

# Printing the confusion matrix of the classifier
cm2 = confusion_matrix(y_test2, y_pred2)
print(cm2)

# The above steps are repeated till each instance in the data has appeared once in the test set

test2=y_test2.index.to_list()

test2.extend(test1)
y_test3=y.drop(test2,axis=0).sample(n=1386)

y_train3=y.drop(y_test3.index.to_list(),axis=0)

X_train3=X.loc[y_train3.index.to_list()]

X_test3=X.loc[y_test3.index.to_list()]

logisticRegr.fit(X_train3, y_train3)

y_pred3=logisticRegr.predict(X_test3)

print("accuracy_score",accuracy_score(y_test3,y_pred3))

cm3 = confusion_matrix(y_test3, y_pred3)
print(cm3)

test3=y_test3.index.to_list()

test3.extend(test2)


y_test4=y.drop(test3,axis=0).sample(n=462)

y_train4=y.drop(y_test4.index.to_list(),axis=0)

X_train4=X.loc[y_train4.index.to_list()]

X_test4=X.loc[y_test4.index.to_list()]

logisticRegr.fit(X_train4, y_train4)

y_pred4=logisticRegr.predict(X_test4)

print("accuracy_score",accuracy_score(y_test4,y_pred4))

cm4 = confusion_matrix(y_test4, y_pred4)
print(cm4)
```

# 3.1 Logistic Regression

```python
# Resampling the data to get a suffled dataset
df=df.sample(frac=1).copy()

# Considering all the features except P6_Ratio as dependent features
X=df.iloc[:,1:-1].drop(["P6_Ratio"],axis=1)

# Taking P6_Ratio as target variable
y=df["P6_Ratio"]

# Splitting into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)

# Creating an instance of logistic regression model
logisticRegr = LogisticRegression(solver='lbfgs')

# Fitting the model for training samples
logisticRegr.fit(X_train, y_train)

# Predicting for test samples 
y_pred=logisticRegr.predict(X_test)

# Printing the accuracy score of the model
print("accuracy_score",accuracy_score(y_test,y_pred))

logisticRegr.score(X_test, y_test)

# Obtaining confusion matrix of the classifier
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Collecing the indices of customer companies which have participated in test set
test1=y_test.index.to_list()

# Dropping the companies which have participated in test set so that they can be replaced with fresh set of samples in test output
# A fresh set of 1386 samples are selected at random and are taken as next test set
y_test2=y.drop(test1,axis=0).sample(n=1386)

# The samples excluding from the new test set are taken into training output
y_train2=y.drop(y_test2.index.to_list(),axis=0)

# The independent features corresponding to the new training output are taken into new training set of independent features
X_train2=X.loc[y_train2.index.to_list()]

# The independent features corresponding to the new test output are taken into new test set of independent features
X_test2=X.loc[y_test2.index.to_list()]

# Fitting model with new set of training and test data
logisticRegr.fit(X_train2, y_train2)

# Predicting for target variable in test set
y_pred2=logisticRegr.predict(X_test2)

# Printing the accuracy of the model
print("accuracy_score",accuracy_score(y_test2,y_pred2))

# Printing the confusion matrix of the classifier
cm2 = confusion_matrix(y_test2, y_pred2)
print(cm2)

# The above steps are repeated till each instance in the data has appeared once in the test set

test2=y_test2.index.to_list()

test2.extend(test1)
y_test3=y.drop(test2,axis=0).sample(n=1386)

y_train3=y.drop(y_test3.index.to_list(),axis=0)

X_train3=X.loc[y_train3.index.to_list()]

X_test3=X.loc[y_test3.index.to_list()]

logisticRegr.fit(X_train3, y_train3)

y_pred3=logisticRegr.predict(X_test3)

print("accuracy_score",accuracy_score(y_test3,y_pred3))

cm3 = confusion_matrix(y_test3, y_pred3)
print(cm3)

test3=y_test3.index.to_list()

test3.extend(test2)


y_test4=y.drop(test3,axis=0).sample(n=462)

y_train4=y.drop(y_test4.index.to_list(),axis=0)

X_train4=X.loc[y_train4.index.to_list()]

X_test4=X.loc[y_test4.index.to_list()]

logisticRegr.fit(X_train4, y_train4)

y_pred4=logisticRegr.predict(X_test4)

print("accuracy_score",accuracy_score(y_test4,y_pred4))

cm4 = confusion_matrix(y_test4, y_pred4)
print(cm4)
```
